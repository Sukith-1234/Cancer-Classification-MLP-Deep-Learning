```python
X=df.drop('benign_0__mal_1',axis=1).values
y=df['benign_0__mal_1'].values
```

```python
# split the data set into  trian and test (0.3)
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
# scaling the train data set
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model=Sequential()

model.add(Dense(30,activation='relu')) # input layer (input has 19 variables hence we use 19 neurones in input layer)
model.add(Dense(15,activation='relu')) # hiiden layer 01

model.add(Dense(1,activation='sigmoid')) # output layer (Binary classification problem hence we use sigmoid fn)
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
loss_df=pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
model=Sequential()

model.add(Dense(30,activation='relu')) # input layer (input has 19 variables hence we use 19 neurones in input layer)
model.add(Dense(15,activation='relu')) # hiiden layer 01

model.add(Dense(1,activation='sigmoid')) # output layer (Binary classification problem hence we use sigmoid fn)
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
# adding early stop for avoid overfitting
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
loss_df=pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
# adding droput layer for avoid overfitting
from tensorflow.keras.layers import Dropout
```

```python
model=Sequential()

model.add(Dense(30,activation='relu')) # input layer (input has 19 variables hence we use 19 neurones in input layer)
model.add(Dropout(0.5)) # dropout layer

model.add(Dense(15,activation='relu')) # hidden layer
model.add(Dropout(0.5))# dropout layer

model.add(Dense(1,activation='sigmoid')) # output layer (Binary classification problem hence we use sigmoid fn)
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
loss_df=pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
y_pred = model.predict(X_test)
y_pred = np.round(y_pred).astype(int)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_true=y_test,y_pred=y_pred))
```

```python
print(confusion_matrix(y_true=y_test,y_pred=y_pred))
```
